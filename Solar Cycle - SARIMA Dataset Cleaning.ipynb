```python
df1 = [fifteen[['Global PSP (vent/cor) [W/m^2]', 'time']], sixteen[['Global PSP (vent/cor) [W/m^2]', 'time']], seventeen[['Global PSP (vent/cor) [W/m^2]', 'time']], eighteen[['Global PSP (vent/cor) [W/m^2]', 'time']], nineteen[['Global PSP (vent/cor) [W/m^2]', 'time']]]
solarGPSPC = pd.concat(df1)
```

```python
solarGPSPC = solarGPSPC.rename(columns = {'Global PSP (vent/cor) [W/m^2]': 'Global PSP [W/m^2]'})
```

```python
solarGPSPC
```

```python
df2 = [solarGPSP, solarGPSPC]
solar_complete = pd.concat(df2)
```

```python
solar_complete = solar_complete[(solar_complete['Global PSP [W/m^2]'] != -99999) & (solar_complete['Global PSP [W/m^2]'] <= 3000)]
solar_complete.reset_index(drop = True)
```

```python
s = pd.to_datetime(min(solar_complete.time))
e = pd.to_datetime(max(solar_complete.time))
solar_complete['time'] = pd.to_datetime(solar_complete.time)
data_res = pd.DataFrame(pd.date_range(s, e, freq = 'H'), columns = ['time'])
```

```python
data_joined = pd.merge(data_res, solar_complete, on=['time'], how = 'left')
data_joined.set_index('time').resample('1H').mean().interpolate()
```

## Discussion of everything done up until this point

1. The EDA was rather straighforward: simple table checking for a few different reasons
    1. Identify the feature with the most entries, as this model only uses a single feature
    2. Identify any obvious outliers 
2. Outlier Handling: values = -999999 and values > 3000 were found to be from systemic issues with the data gathering equipment
3. Since there is only one feature needed per model, the EDA was ended after the feature identification, handling of the outliers, and combining the dataframes
4. Data Cleaning: on top of the already mentioned outlier handling, the time differences between all of the datasets included 1 hour, 5 minutes, and 1 minute intervals. I ensured that the data was all on the same time interval. Also, for missing data, I used a Interpolate Fill Resample method. The data is fitted between the final time value of one dataset, and the first time value of the other, and a point is added at the specified time interval (1H in this case)
5. Limitations of this model: I am assuming that 'Global PSP' (unvented) and 'Global PSP' (vent/cor) values are the same output, and therefore can be treated as 'continuous' of one another. However, considering that there is a decent chance I am wrong about this, I have followed up with the group that gathered the data for more information. The links containing the instrument specs I need to be sure are dead, and if they provide me with the information about the specs I need, updating the model will be super easy as I already have the formulas/process for the fix created
6. Future work (after PSP data) will include running other features through the same process and comparing the results, ensuring that 'Global PSP' vented/unvented can be treated as the same, and updating the model if it isn't. Another idea could be to ignore the hour-interval data, and work with 5 minute intervals, and compare the accuracy of the models then
